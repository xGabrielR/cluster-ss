### Silhouette Plots

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from cluster_ss import plot_silhouette, plot_silhouette_score

cluster_list = [3, 4, 5]
X, y = make_blobs(n_samples=7_000, centers=4, n_features=5)
X = pd.DataFrame(X)
```

```python
axes, fig = plot_silhouette(estimator=KMeans(), X=X, cluster_list=cluster_list)
```

```python
labels_at_k, sil_scores_at_k, _, _ = plot_silhouette_score(estimator=KMeans(), X=X, cluster_list=cluster_list)
```

```python
print(labels_at_k)
```

```python
print(sil_scores_at_k)
```

### Models Fits, Extra Plots & Setup Parameters

```python
from cluster_ss import ClusterSupport

# Add Extra Params
extra_parameters = [
    {'KMeans': {'max_iter': 320,
                'tol': 0.001}},
    {'GaussianMixture': {'max_iter': 110,
                         'n_init': 2,
                         'reg_covar': 1e-05,
                         'tol': 0.01,}}
]

# A List of Clusters
cluster_list = [2, 3, 4, 5]

cs = ClusterSupport(verbose=True, extra_parameters=extra_parameters)
```

```python
# Multiple Silhouettes with Extra params on KMeans and GMM.
axes, fig = cs.plot_multiple_silhouette(cluster_list=cluster_list, fit=True, X=X)
```

```python
silhouettes, no_k_silhouettes, sils_infos_results = cs.fit(X=X, cluster_list=cluster_list, estimators_selection='all')
```
